```python
import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt

```

```python
df = sns.load_dataset('tips')
df.head()
```

# Distribution of Total_bill

```python
Total_bill_Hist = df['total_bill'].mean()
Total_bill_Hist
```

```python
sns.distplot(df['total_bill'])
```

```python
sns.swarmplot(x="sex", y="total_bill", data=df)
```

```python
# Set context to `"paper"`
sns.set_context("paper")
sns.swarmplot(x="sex", y="total_bill", data=df)
```

```python
# Set up a factorplot

# Set context to `"paper"`
sns.set_context("paper")

g = sns.factorplot("day", "tip", "sex", data=df, kind="bar", palette="muted", legend=False)
                   
# Show plot
plt.show()
```
